## Transformation

```python
world_data = df.head(1)
```

```python
df = df.iloc[1:, :]
```

```python
# check for null values

df.isnull().sum()
```

```python
df[pd.isna(df['1990'])]
```

```python
mean_value = df.iloc[111, 5:].mean().round(2)
print(mean_value)
```

```python
# replace the null with mean value

df.at[112, '1990'] = mean_value
```

```python
df.isnull().sum()
```

Recode

```python
# add new column

df['average_emissions'] = df.iloc[:, 5:].mean(axis=1)
```
